```python
import geopy.distance
```

```python
merged.columns
```

```python
testdf = merged[merged['COMMAREA']==2]
testdf
```

```python
def fun(test):
  def f(x):
    cord1 = (float(x['lat']),float(x['lon']))
    cord2 = (float(test.Latitude),float(test.Longitude))
    distance = geopy.distance.geodesic(cord1, cord2).miles
    distances.append(distance)
    races.append(x[13:20])

  distances = []
  races = []
  testdf.apply(f,axis = 1)
  racedf = pd.DataFrame(races)
  racedf['Distance'] = distances
  means = []
  means.append(racedf[racedf['White']!=0]['Distance'].mean())
  means.append(racedf[racedf['Asian']!=0]['Distance'].mean())
  means.append(racedf[racedf['Black']!=0]['Distance'].mean())
  means.append(racedf[racedf['Hispanic']!=0]['Distance'].mean())
  means.append(racedf[racedf['Pacific Islander']!=0]['Distance'].mean())
  means.append(racedf[racedf['Native']!=0]['Distance'].mean())
  means.append(racedf[racedf['Other']!=0]['Distance'].mean())
  return means
```
