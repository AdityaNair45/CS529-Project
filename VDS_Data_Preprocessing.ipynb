```python
sample = pd.read_csv('IL_dataset_1.csv')
sample.head()
```

```python
sample['State Code'] = sample.FIPS.apply(lambda x: str(x)[:2])
sample['County Code'] = sample.FIPS.apply(lambda x: str(x)[2:5])
sample['Tract Code'] = sample.FIPS.apply(lambda x: str(x)[5:11])
sample['Block Code'] = sample.FIPS.apply(lambda x: str(x)[11:15])
```

```python
sample.head()
```

```python
area = pd.read_csv('CensusTractsTIGER2010.csv')
```

```python
area.head()
```

```python
area['Tract Code'] = area['Tract Code'].astype(str)
```

```python
area.head(20)
```

```python
area['Tract Code'] = area['Tract Code'].str.zfill(6)
```

```python
area.head(20)
```

```python
final = sample.merge(area[['Tract Code','COMMAREA']],how='inner', left_on='Tract Code', right_on='Tract Code')
```

```python
final.head(20)
```

```python
final.to_csv('Final.csv')
```
