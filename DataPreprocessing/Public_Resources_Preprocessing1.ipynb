```python
from shapely.geometry import shape, Point
import shapefile
```

```python
r = shapefile.Reader("geo_export_857e7ab3-ee77-40cd-b5e0-5222e9567269.shp")

# get the shapes
shapes = r.shapes()

# build a shapely polygon from your shape
polygon = []
for x in range(0,len(r)):
  polygon.append(shape(shapes[x])) 

list = []
for y in hospital.index:
  for x in range(0,len(r)):
    if polygon[x].contains(Point(hospital.Longitude[y],hospital.Latitude[y])):
      list.append(comm[x])

len(list)
```

```python
hospital['Comm_Area'] = list
```

```python
hospital.head()
```

```python
hospital.to_csv('Hospitals_Processed.csv')
```
