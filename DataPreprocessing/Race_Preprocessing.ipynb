```python
import pandas as pd
```

```python
race = pd.read_csv('Race_Category.csv')
```

```python
race.columns = ['Code','Value']
#,'White','Asian','Pacific Islander','Black','Hispanic','Native','Other']
```

```python
race.head()
```

```python
sample = pd.read_csv('Final.csv')
```

```python
sample.head()
```

```python
merged = sample.merge(race,how='left',left_on='race',right_on='Code')
```

```python
merged.head(20)
```

```python
merged['White'] = merged.Value.apply(lambda x : 1 if ('White' in str(x)) else 0)
```

```python
merged.head()
```

```python
merged['Asian'] = merged.Value.apply(lambda x : 1 if ('Asian' in str(x)) else 0)
merged['Black'] = merged.Value.apply(lambda x : 1 if ('Black' in str(x)) else 0)
merged['Hispanic'] = merged.Value.apply(lambda x : 1 if ('Hispanic' in str(x)) else 0)
merged['Pacific Islander'] = merged.Value.apply(lambda x : 1 if ('Pacific' in str(x)) else 0)
merged['Native'] = merged.Value.apply(lambda x : 1 if ('Alaska' in str(x)) else 0)
merged['Other'] = merged.Value.apply(lambda x : 1 if ('Some Other Race' in str(x)) else 0)
```

```python
race_final = merged.groupby('COMMAREA')['White','Asian','Black','Hispanic','Pacific Islander','Native','Other'].sum()
```

```python
race_final.to_csv('CommArea_Race.csv')
```
